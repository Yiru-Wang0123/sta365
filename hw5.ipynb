### Q2: Continue "Q2" of the previous weeks 3 and 4

1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt

n = 30
true_theta = 0
true_tau = 1
np.random.seed(365)
x = np.random.normal(loc=true_theta, scale=1/np.sqrt(true_tau), size=n)

def run_pymc_model(model_name, model):
    with model:
        trace = pm.sample(2000, tune=1000, return_inferencedata=True)
        az.plot_trace(trace)
        plt.suptitle(f'Trace Plots for {model_name}')
        plt.show()
        
        az.plot_posterior(trace)
        plt.suptitle(f'Posterior Distributions for {model_name}')
        plt.show()
        
        print(f'\nSummary for {model_name}:')
        print(az.summary(trace))
```

1. Normal prior for theta, Gamma prior for tau

```python
with pm.Model() as model_1:
    theta = pm.Normal("theta", mu=0, sigma=1)
    tau = pm.Gamma("tau", alpha=2, beta=2)
    likelihood = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)

run_pymc_model("Normal-Theta, Gamma-Tau", model_1)
```

2. Non-normal prior for theta, Non-gamma prior for tau (e.g., Laplace and HalfNormal)

```python
with pm.Model() as model_2:
    theta = pm.Laplace("theta", mu=0, b=1)
    tau = pm.HalfNormal("tau", sigma=1)
    likelihood = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)

run_pymc_model("Laplace-Theta, HalfNormal-Tau", model_2)
```

3. Another non-normal prior for theta, another non-gamma prior for tau (e.g., Uniform and Exponential)

```python
with pm.Model() as model_3:
    theta = pm.Uniform("theta", lower=-5, upper=5)
    tau = pm.Exponential("tau", lam=1)
    likelihood = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)

run_pymc_model("Uniform-Theta, Exponential-Tau", model_3)
```

### Q3: Slice Sampling

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant. In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*


- The curve that we are sampling beneath is the pdf of the target distribution. In the code, it is $f(x) = Beta(1.5, 3)$. As `f = lambda x: stats.beta(1.5, 3).pdf(x)`.  It describes the relative likelihood of different values of $x$. Slice sampling is used to draw samples from this distribution by sampling uniformly under the curve $f(x)$, which ensures that we draw samples proportaional to this density.
- The initial value for $x$ in the code is set to 0.25. As `x[:2, 0] = 0.25`. It can be chosen randomly or based on prior knowledge.
- To create the draws for Gibbs Sampling, we need to
    1. initialize the the chain with an initival value for $x$, which in this case 0.25
    2. iterate through the gibbs sampling loop. For each iteration $t$, update the variables using their full conditional distributions. In this case, we're using slice sampling to draw from the full conditional distribution of $x$
    3. For the slice sampling, we need to:
        - Draw a slice level $y$ uniformly from the interval $[0, f(x_t)]$. `x[t, 1] = stats.uniform().rvs() * f(x[t, 0])`
        - Find the inveral $[a, b]$ where the curve $f(x)$ is above the slice level $y$. This is done by checking the grid point `a, b = x_grid[f(x_grid) > y][[0, -1]] + [-x_grid_delta, x_grid_delta]`. It is also is extended slightly to ensure the endpoints satisfy $f(a)<yf(a)<y$ and $f(b)<yf(b)<y$.
        - Sample the next point $x_{t+1}$ uniformly from the interval $[a, b]$. `a, b = x_grid[f(x_grid) > y][[0, -1]] + [-x_grid_delta, x_grid_delta]`
        - If this `f(x_{t+1}) > y`, accept $x_{t+1}$ as the new sample. Otherwise, shrink the interval and try again `x_ = a + stats.uniform().rvs() * (b - a)`
- Repeat for the next iteration of $t$ in Gibbs sampling, and updating $x$ using slice sampling
